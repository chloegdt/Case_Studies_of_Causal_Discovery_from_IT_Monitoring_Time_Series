```python
import pandas as pd
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

# Antivirus

## Antivirus 1

```python
antivirus_activity_1 = pd.read_csv('data/Antivirus_Activity/preprocessed_1.csv', parse_dates=True) 
antivirus_activity_1['timestamp'] = pd.to_datetime(antivirus_activity_1['timestamp'])
antivirus_activity_1 = antivirus_activity_1.set_index('timestamp')
antivirus_activity_1.info()
```

```python
plt.figure(figsize=(200, 400))

labels = ['RP',
          'MUGV',
          'CUGV',
          'CUP',
          'RV',
          'CUGP',
          'CUV',
          'ChP',
          'T',
          'MUGP',
          'MUP',
          'ChIE',
          'MUV'
         ]

axes = antivirus_activity_1[[
    '0_C_read_Portal',
    'Physical_Memory_prct_used_VDI',
    'cpu_prct_used_VDI',
    'cpu_usage_Portal',
    '0_C_read_VDI',
    'cpu_prct_used_Portal',
    'cpu_usage_VDI',
    'Chargement_portail',
    'Default_Transaction',
    'Physical_Memory_prct_used_Portal',
    'memory_usage_Portal',
    'Chargement_IE',
    'memory_usage_VDI',
    ]].plot(subplots=True);

for axe, label in zip(axes, labels):
    axe.legend([label])

directory = 'plots'
plot = 'antivirus1.png'
path = os.path.join(directory, plot)

os.makedirs(directory, exist_ok=True)
plt.savefig(path) 
```
