## MoM - graphe causal à partir du plot

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_edges_from([
    ('Pub', 'Mes'),
    ('Con', 'Mes'),
    ('Pub', 'DiskR'),
    ('Pub', 'DiskW'),
    ('Pub', 'RAM'),
    ('Mes', 'DiskR'),
    ('Mes', 'DiskW'),
    ('Mes', 'RAM')
])

G.add_node('CPU')

pos = nx.spring_layout(G, seed=42)
pos['Mes'] = [0.5, 0.8]
pos['Pub'] = [0.2, 0.8]
pos['Con'] = [0.4, 0.6]
pos['DiskR'] = [0.6, 0.4]
pos['DiskW'] = [0.8, 0.4]
pos['RAM'] = [0.7, 0.6]
pos['CPU'] = [0.1, 0.1]

# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrowsize=20)
plt.title('Causal Graph MoM1')
plt.show()
```

## Ingestion - Uniquement sur les variables et le contexte

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the nodes
nodes = ["PMDB", "MDB", "CMB", "MB", "LMB", "RTMB", "GSIB", "ESB"]

# Define the edges based on the causal relationships
edges = [
    ("PMDB", "MDB"),
    ("PMDB", "CMB"),
    ("MDB", "RTMB"),
    ("MDB", "ESB"),
    ("CMB", "MB"),
    ("CMB", "LMB"),
    ("CMB", "RTMB"),
    ("MB", "LMB"),
    ("MB", "GSIB"),
    ("RTMB", "GSIB"),
    ("RTMB", "ESB"),
    ("GSIB", "ESB")
]

# Create a directed graph
G = nx.DiGraph()

# Add nodes and edges to the graph
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Set the layout for the graph
pos = nx.spring_layout(G)

# Draw the graph
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=12, font_weight="bold", arrows=True, arrowsize=20)
plt.title("Causal Graph of Ingestion Dataset", size=15)
plt.show()
```

## Antivirus - Contexte & Variables

```python
import matplotlib.pyplot as plt
import networkx as nx

nodes_antivirus = ["CUV", "CUGV", "MUV", "MUGV", "RV", "ChIE", "CUP", "CUGP", "MUP", "MUGP", "RP", "ChP", "T"]

# Define the edges based on the identified causal relationships
edges_antivirus = [
    ("CUV", "CUGV"),
    ("CUGV", "ChIE"),
    ("CUGV", "RV"),
    ("MUV", "MUGV"),
    ("MUGV", "ChIE"),
    ("MUGV", "T"),
    ("RV", "ChIE"),
    ("ChIE", "T"),
    ("CUP", "CUGP"),
    ("CUGP", "ChP"),
    ("MUP", "MUGP"),
    ("MUGP", "ChP"),
    ("MUGP", "T"),
    ("RP", "ChP"),
    ("RP", "T"),
    ("ChP", "T")
]

# Create a directed graph for the antivirus server activity
G_antivirus = nx.DiGraph()

# Add nodes and edges to the graph
G_antivirus.add_nodes_from(nodes_antivirus)
G_antivirus.add_edges_from(edges_antivirus)

# Set the layout for the graph
pos_antivirus = nx.spring_layout(G_antivirus, seed=42)  # fixed layout for reproducibility

# Draw the graph
plt.figure(figsize=(16, 12))
nx.draw(G_antivirus, pos_antivirus, with_labels=True, node_size=3000, node_color="lightblue", font_size=12, font_weight="bold", arrows=True, arrowsize=20)
plt.title("Causal Graph of Antivirus Server Activity Dataset", size=15)
plt.show()
```
