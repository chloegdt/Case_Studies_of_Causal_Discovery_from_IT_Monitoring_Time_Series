# Graphes causaux (générés par CHATGPT à partir du contexte & nom de variables)

## MoM 

```python
import os
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()

edges = [
    ("Pub", "Mes"),
    ("Pub", "Con"),
    ("Pub", "DiskW"),
    ("Pub", "DiskR"),
    ("Pub", "Cpu"),
    ("Pub", "Ram"),
    ("Con", "Mes"),
    ("Con", "Cpu"),
    ("Con", "Ram"),
    ("Con", "DiskW"),
    ("Con", "DiskR"),
    ("Mes", "Cpu"),
    ("Mes", "DiskR"),
    ("Cpu", "Ram"),
    ("Ram", "DiskR"),
    ("Ram", "DiskW"),
]

G.add_edges_from(edges)


pos = nx.spring_layout(G)

# Créer le dossier 'mes_graphes' s'il n'existe pas
output_dir = 'mes_graphes'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Nom du fichier de sauvegarde
output_file = os.path.join(output_dir, 'graphe_causal_mom_gpt.png')

# Dessiner le graphe
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", arrows=True, arrowsize=20)
plt.title('Graphe Causal du Dataset Mom')

plt.savefig(output_file)

plt.show()
```
